```python
import pandas as pd
```

```python
# Read the excel file
filepath = 'src/data/'
input_file_name = 'first_table.xlsx'
df_original = pd.read_excel('{0}{1}'.format(filepath, input_file_name), sheet_name=None)
```

```python
df = df_original.copy()
```

```python
# Get the sheet names
sheet_names = df.keys()
for elements in sheet_names:
    print(elements)
```

```python
worksheet = df['Page003']
worksheet.head()
```

```python
#rename columns to row 1
worksheet.columns = worksheet.iloc[2]
worksheet.head()
```

```python
#drop row 0
worksheet = worksheet.drop([0])
#drop row 1
worksheet = worksheet.drop([1])
#drop row 2
worksheet = worksheet.drop([2])
#drop row  43
worksheet = worksheet.drop([43])
#drop row  44
worksheet = worksheet.drop([44])
```

```python
worksheet.head()
```

```python
#get column names
column_names = worksheet.columns
#print column names with index
for index, element in enumerate(column_names):
    print(index, element)
```

```python
#drop last column
worksheet = worksheet.drop(labels=[column_names[9]], axis=1)
worksheet.head()
```

```python
#rename column 0 into Area
worksheet = worksheet.rename(columns={column_names[0]:'Area'})
```

# WORKING CODE

```python
worksheet.reset_index(level='Gender')
```

```python
#drop index
worksheet = worksheet.reset_index(drop=True)
worksheet.head()
```

```python
#rename index into ID
worksheet = worksheet.rename_axis('ID').reset_index()
worksheet.head()
```

```python
#get column names
column_names = worksheet.columns
#print column names with index
for index, element in enumerate(column_names):
    print(index, element)
```

```python
#rename index column into indexID
worksheet = worksheet.rename(columns={column_names[0]:'indexID'})
```

```python
input_file_name = 'second_table.xlsx'
with pd.ExcelWriter('src/data/{0}'.format(input_file_name)) as writer:
    worksheet.to_excel(writer, sheet_name='all_home_type')
```
